Este código é um script Python projetado para coletar informações de usuários e oferecer recomendações de cursos. Ele utiliza a biblioteca pandas para manipulação de dados e a API google.generativeai para gerar sugestões personalizadas.

O script começa importando as bibliotecas necessárias e define o caminho de um arquivo Excel, onde os dados serão armazenados. Em seguida, tenta carregar os dados desse arquivo; caso ele não exista, um DataFrame vazio é criado. A chave da API do Google é obtida a partir dos dados do usuário, permitindo que a biblioteca de IA seja configurada para gerar conteúdo.

A função principal do código apresenta um menu interativo com três opções: registrar informações do usuário, obter recomendações de cursos ou sair. Ao registrar, o usuário insere dados como nome, e-mail e interesses, que são armazenados no DataFrame e salvos no arquivo Excel. Para as recomendações, o script gera um prompt baseado nos interesses do usuário e utiliza a IA para sugerir cursos relevantes. O código continua em loop até que o usuário decida sair, garantindo uma experiência interativa e dinâmica.

```
!pip install -U google-generativeai
import pandas as pd
import google.generativeai as genai
from google.colab import userdata
```

```
file_path = '/content/drive/MyDrive/Base de dados.xlsx'

try:
    data = pd.read_excel(file_path)
except FileNotFoundError:
    data = pd.DataFrame()

GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

def add_user_info():
    user_info = {
        'ID USUÁRIO': input("Digite o seu ID: "),
        'NOME': input("Digite o seu nome: "),
        'E-MAIL': input("Digite seu e-mail: "),
        'SENHA': input("Digite sua senha: "),
        'DATA NASCIMENTO': input("Digite sua data de nascimento (YYYY-MM-DD): "),
        'CPF': input("Digite seu CPF: "),
        'PROFISSÃO': input("Digite sua profissão: "),
        'CATEGORIA DO CURSO': input("Digite a categoria do curso realizado: "),
        'TITULO': input("Digite o título do curso: "),
        'INTERESSE PROFISSIONAL': input("Digite seu interesse profissional: "),
        'INTERESSE PESSOAL': input("Digite seu interesse pessoal: "),
        'AREA INTERESSE': input("Digite sua área de interesse: "),
        'NOTA': float(input("Digite a nota do curso: ")),
        'COMENTARIO': input("Digite um comentário sobre o curso: ")
    }

    global data
    new_entry = pd.DataFrame([user_info])
    data = pd.concat([data, new_entry], ignore_index=True)
    data.to_excel(file_path, index=False)
    print("Informações do usuário registradas com sucesso!")

def generate_course_recommendations(interesse_profissional, interesse_pessoal, area_interesse):
    prompt = (
        f"Baseado nos seguintes critérios: \n"
        f"- Interesses profissionais: {interesse_profissional}\n"
        f"- Interesses pessoais: {interesse_pessoal}\n"
        f"- Área de interesse: {area_interesse}\n\n"
        "Sugira uma lista de 3 a 5 cursos relevantes. Para cada curso, forneça o nome e uma breve descrição."
        " Limite a resposta a no máximo 250 palavras."
    )

    model = genai.GenerativeModel('gemini-pro')
    response = model.generate_content(prompt)

    return response.text

def main():
    while True:
        print("\nEscolha uma opção:")
        print("1. Obter recomendações de cursos")
        print("2. Registrar informações do usuário")
        print("3. Sair")

        choice = input("Digite o número da opção desejada: ")

        if choice == '1':
            interesse_profissional = input("Digite seu interesse profissional: ")
            interesse_pessoal = input("Digite seu interesse pessoal: ")
            area_interesse = input("Digite sua área de interesse: ")

            generated_recommendations = generate_course_recommendations(
                interesse_profissional, interesse_pessoal, area_interesse
            )
            print(f"\nRecomendações de cursos:\n{generated_recommendations}")

        elif choice == '2':
            add_user_info()
        elif choice == '3':
            print("Saindo...")
            break

        else:
            print("Opção inválida. Tente novamente.")

if __name__ == "__main__":
    main()
```
